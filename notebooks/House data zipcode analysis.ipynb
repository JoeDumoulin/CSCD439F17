```python
import pandas as pd
df = pd.read_csv('kc_house_data.csv')
df.head()

```

```python
df.groupby('zipcode').count().shape
```

```python
dfbyzip = df[['zipcode', 'price', 'bedrooms', 'sqft_living', 'yr_built']].groupby('zipcode').mean()
dfbyzip.head()
```

```python
import math
dfbyzip = df[['zipcode', 'price', 'bedrooms', 'sqft_living', 'yr_built']].groupby('zipcode').mean() \
    .agg({ 'price': int, 'bedrooms':int, 'sqft_living':int, 'yr_built':int})
dfbyzip
```
